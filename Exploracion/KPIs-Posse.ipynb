```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('D:\data\Proyectos\Grupal\DS-Proyecto_Grupal_Olist\data\e-comerce_Olist_dataset\olist_order_reviews_dataset.csv')

buenas=  df[(df['review_score'] == 4) | (df['review_score'] == 5)]

buenas['review_answer_timestamp'] = pd.to_datetime(buenas['review_answer_timestamp'])
buenas['review_creation_date'] = pd.to_datetime(buenas['review_creation_date'])

df['review_creation_date'] = pd.to_datetime(df['review_creation_date'])
df['review_answer_timestamp'] = pd.to_datetime(df['review_answer_timestamp'])

kpi_entregasbuenas = buenas.groupby(buenas['review_creation_date'].dt.year)['review_score'].count() / df.groupby(df['review_creation_date'].dt.year)['review_score'].count() * 100

colores=['blue','green','orange','yellow','cyan']

kpi_entregasbuenas.plot(kind='bar', color = colores, ylabel = 'Porcentaje de buenas entregas', xlabel = 'Años',
                        figsize = (20,10)
)
plt.title('Crecimiento en la mejora de entregas a traves de los años')
plt.show()
```
